```python
import requests
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Access the variables
X_N8N_API_KEY = os.getenv("X_N8N_API_KEY")
N8N_HOST = os.getenv("N8N_HOST")
N8N_API_VERSION = os.getenv("N8N_API_VERSION")
headers = {
    "accept": "application/json",
    "X-N8N-API-KEY": X_N8N_API_KEY,
}
```

```python
# Get All Workflows
url = f"{N8N_HOST}/api/{N8N_API_VERSION}/workflows?active=false"
response = requests.get(url, headers=headers)
print(response.json())
```

```python
front_desk_agent_workflow_id = "029b0SUSmpcOCby8"
# Get All Execution of a Workflow
response_list = []
nextCursor = ""
while True:
    if nextCursor == "":
        url = f"{N8N_HOST}/api/{N8N_API_VERSION}/executions?workflowId={front_desk_agent_workflow_id}&&limit=1&&includeData=true"
    else:
        url = f"{N8N_HOST}/api/{N8N_API_VERSION}/executions?workflowId={front_desk_agent_workflow_id}&&limit=1&&includeData=true&&cursor={nextCursor}"
    response = requests.get(url, headers=headers)
    response_list.append(response.json())
    nextCursor = response.json()['nextCursor']
    if nextCursor is None:
        break
    if len(response_list) > 10:
        break
```

```python
# load json and dump to csv
import json
import csv
json_file = './100_short_mockup_question_gpt-o3-20250207.json'
csv_file = './100_short_mockup_question_gpt-o3-20250207.csv'
with open(json_file, 'r') as f:
    data = json.load(f)
    with open(csv_file, 'w') as f:
        csv_writer = csv.writer(f)
        # Write the header
        csv_writer.writerow(data[0].keys())
        # Write the data
        for row in data:
            csv_writer.writerow(row.values())
```
